```python
df1 = run_time(75, probabilities)
df2 = run_time(100, probabilities)
df3 = run_time(125, probabilities)
df4 = run_time(150, probabilities)

results = [df1, df2, df3, df4]
result = []

for dataframe in results:
    
    edps_runtimes = dataframe['Edps']
    one_runtimes = dataframe['One Tree']
    multi_runtimes = dataframe['Multiple Trees']
    number_nodes = dataframe['Number of nodes'][0]
    
    for i in range(len(edps_runtimes)):
        result.append((number_nodes, edps_runtimes[i], "Edps"))
        result.append((number_nodes, one_runtimes[i], "One Tree"))
        result.append((number_nodes, multi_runtimes[i], "Multiple Trees"))
    
for j in ["Edps", "One Tree", "Multiple Trees"]:
    a = [element[0] for element in result if element[2] == j]
    b = [element[1] for element in result if element[2] == j]
    plt.scatter(a, b, label= j)

plt.xlabel('Number of Nodes')
plt.ylabel('Runtime in s')
plt.legend()
plt.show()

```

# Korrelation between the number of the edges and Runtime

```python
def korrelation_runtime_edges():
    
    dataframes = []
    for node_number in range(100,500, 100):
        print(f'Node numberrrrrrrrrrrrrrrrrrrrrrrrrr:  {node_number}')
        p_list = [0.06, 0.5, 0.6, 0.8, 1]
        
        for proba in p_list:
            graph = nx.erdos_renyi_graph(node_number, proba)
            
            source, destination = None, None
            
            while source == destination:
                source = random.choice(list(graph.nodes()))
                destination = random.choice(list(graph.nodes()))
                
            size_of_graph = graph.number_of_nodes()
            probability = proba
            edp_times = []
            one_times = []
            multi_times = []
            
            for j in range(10):
                
                edp_time = timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="edps"), number=1)

                one_time = timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="onetree"), number=1)

                multi_time = timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="multitree"), number=1)
                
                print(f'{edp_time}')
                print(f'{one_time}')
                print(f'{multi_time}')
                
                edp_times.append(edp_time)
                one_times.append(one_time)
                multi_times.append(multi_time)
                
                df = pd.DataFrame({
                    "Source node": source,
                    "Destination node": destination,
                    "Number of nodes": size_of_graph,
                    "Number of edges": graph.number_of_edges(),
                    "Probability": probability,
                    "Edps": edp_times,
                    "One Tree": one_times,
                    "Multiple Trees": multi_times,
                })
                
                dataframes.append(df)
                
    my_dataframe = pd.concat(dataframes)
#     result = my_dataframe.groupby(["Number of edges"]).mean().reset_index()
    return my_dataframe
    
#     plt.plot(result["Number of edges"], result["Edps"], label="Edge disjoint paths")
#     plt.plot(result["Number of edges"], result["One Tree"], label="One Tree")
#     plt.plot(result["Number of edges"], result["Multiple Trees"], label="Multiple Trees")
#     plt.xlabel("Number of edges") 
#     plt.ylabel("Runtime in ms")
#     plt.title("Korrelation between number of edges and runtime")
#     plt.savefig('kor.svg')
                
        
```

```python
result = korrelation_runtime_edges()
result.to_csv('korrelation_edges_runtime.csv', index=False)
print(result.to_string())
```

```python
results = result.groupby(["Number of edges"]).mean().reset_index()
plt.plot(results["Number of edges"], results["Edps"], label="Edge disjoint paths")
plt.plot(results["Number of edges"], results["One Tree"], label="One Tree")
plt.plot(results["Number of edges"], results["Multiple Trees"], label="Multiple Trees")
plt.xlabel("Number of edges") 
plt.ylabel("Runtime in ms")
plt.title("Korrelation between number of edges and runtime")
plt.savefig('korrelation.svg')
```

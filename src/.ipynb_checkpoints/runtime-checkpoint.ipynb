```python
from Overload import Overload
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import timeit
overload = Overload()
import random
from tabulate import tabulate
```

```python
def korrelation_runtime_edges():
    
    dataframes = []
    for node_number in range(200,300, 100):
        print(f'Node numberrrrrrrrrrrrrrrrrrrrrrrrrr:  {node_number}')
        p_list = [0.06, 0.5, 0.6, 0.8, 1]
        
        for proba in p_list:
            graph = nx.erdos_renyi_graph(node_number, proba)
            
            source, destination = None, None
            
            while source == destination:
                source = random.choice(list(graph.nodes()))
                destination = random.choice(list(graph.nodes()))
                
            size_of_graph = graph.number_of_nodes()
            probability = proba
            edp_times = []
            one_times = []
            multi_times = []
            
            for j in range(10):
                
                edp_time = timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="edps"), number=5)

                one_time = timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="onetree"), number=5)

                multi_time = timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="multitree"), number=10)
                
                print(f'{edp_time}')
                print(f'{one_time}')
                print(f'{multi_time}')
                
                edp_times.append(edp_time)
                one_times.append(one_time)
                multi_times.append(multi_time)
                
                df = pd.DataFrame({
                    "Source node": source,
                    "Destination node": destination,
                    "Number of nodes": size_of_graph,
                    "Number of edges": graph.number_of_edges(),
                    "Probability": probability,
                    "Edps": edp_times,
                    "One Tree": one_times,
                    "Multiple Trees": multi_times,
                })
                
                dataframes.append(df)
                
#     my_dataframe = pd.concat(dataframes)
#     return my_dataframe
    return dataframes
```

```python
result = korrelation_runtime_edges()
print(result.to_string())
```

```python
def korrelation_runtime_edges_up():
    
    dataframes = []
    for node_number in range(200,300, 100):
        print(f'Node numberrrrrrrrrrrrrrrrrrrrrrrrrr:  {node_number}')
        p_list = [0.06, 0.5, 0.6, 0.8, 1]
        
        for proba in p_list:
            graph = nx.erdos_renyi_graph(node_number, proba)
            print(f'Probbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbb {proba}')
            source, destination = None, None
            
            while source == destination:
                source = random.choice(list(graph.nodes()))
                destination = random.choice(list(graph.nodes()))
                
            size_of_graph = graph.number_of_nodes()
            probability = proba
            edp_times = []
            one_times = []
            multi_times = []
            
            for j in range(10):
                
                edp_time = round(timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="edps"), number=1, timer=timeit.default_timer), 3)

                one_time = round(timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="onetree"), number=1, timer=timeit.default_timer), 3)

                multi_time = round(timeit.timeit(
                    lambda: overload.compute_path_for_one_source(graph, source, destination, [], version="multitree"), number=1, timer=timeit.default_timer), 3)
                
                print(f'{edp_time}')
                print(f'{one_time}')
                print(f'{multi_time}')
                
                edp_times.append(edp_time)
                one_times.append(one_time)
                multi_times.append(multi_time)
                
                df = pd.DataFrame({
                    "Source node": source,
                    "Destination node": destination,
                    "Number of nodes": size_of_graph,
                    "Number of edges": graph.number_of_edges(),
                    "Probability": probability,
                    "Edps": edp_times,
                    "One Tree": one_times,
                    "Multiple Trees": multi_times,
                })
                
                dataframes.append(df)
                
#     my_dataframe = pd.concat(dataframes)
#     return my_dataframe
    return dataframes
```

```python
results = two_three.groupby(["Number of edges"]).mean().reset_index()
plt.plot(results["Number of edges"], results["Edps"], label="Edge disjoint paths")
plt.plot(results["Number of edges"], results["One Tree"], label="One Tree")
plt.plot(results["Number of edges"], results["Multiple Trees"], label="Multiple Trees")
plt.xlabel("Number of edges") 
plt.ylabel("Runtime in s")
plt.title("Korrelation between number of edges and runtime")
plt.legend()
plt.savefig('korrelation200300knoten.svg')
```
